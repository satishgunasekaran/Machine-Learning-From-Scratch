```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("Salary_Data.csv")
```

```python
plt.scatter(data['YearsExperience'],data['Salary'])
```

```python
class LinearRegression:
    def __init__(self, lr = 0.01, epochs = 1000):
        self.slope = np.random.randn(1)
        self.intercept = np.random.randn(1)
        self.lr = lr
        self.epochs = epochs
        
    def predict(self, x):
        return self.slope * x + self.intercept
    
    def fit(self, X, y, plot = False):
        for i in range(self.epochs):
            error = 0
            d_slope = 0
            d_intercept = 0
            
            for j in range(len(X)):
                y_pred = self.predict(X[j])
                error += (y_pred - y[j])**2
                d_slope += 2 * (y_pred - y[j]) * X[j]
                d_intercept += 2 * (y_pred - y[j])
                
            error = error / len(X)
            d_slope = d_slope / len(X)
            d_intercept = d_intercept / len(X)
            
            if  i % 100 == 0:
                print(f"Error at {i} epoch is {error}")
                print(f"Slope at {i} epoch is {self.slope}")
                print(f"Intercept at {i} epoch is {self.intercept}")
                
                if plot:
                    plt.scatter(X, y)
                    plt.plot(X, self.predict(X), color = 'red')
                    plt.show() 
            
            self.slope = self.slope - self.lr * d_slope
            self.intercept = self.intercept - self.lr * d_intercept
    
            
            
          
                
```

```python
linreg = LinearRegression()
linreg.fit(data['YearsExperience'], data['Salary'], plot = True)
```
